# Q-2

Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
and make data frame

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
url = 'http://www.imdb.com/chart/top'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
movies = soup.select('td.titleColumn')
crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
ratings = [b.attrs.get('data-value')
        for b in soup.select('td.posterColumn span[name=ir]')]
```

```python
# create a empty list for storing
# movie information
list = []
 
# Iterating over movies to extract
# each movie's details
for index in range(0, 100):
     
    # Separating movie into: 'place','Name','rating','year'
    
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    year = re.search('\((.*?)\)', movie_string).group(1)
    place = movie[:len(str(index))-(len(movie))]
    data = {"place":place,
            "Name": movie_title,
            "rating": ratings[index],
            "year of release": year}
    list.append(data)
```

```python
for movie in list:
    print(movie['place'],'-',movie['Name'],'-' ,movie['rating'],'-',movie['year of release'])
  
```

```python
#saving the list as dataframe
#then converting into .csv file

df = pd.DataFrame(list)
df.to_csv('imdb_top_100_movies.csv',index=False)
```

```python
df = pd.DataFrame(list)
df
```

```python
Image_URL = []

for i in soup.find_all('img',class_='no-img'):
    
    Image_URL.append(i.get('data-src'))
    
Image_URL
```

a)Top 10 ODI teams in men’s cricket along with the records for matches, points and rating

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
Matches=[]

for i in soup.find_all("td",class_='rankings-block__banner--matches'):# first place team ratings
    
    Matches.append(i.text.replace("\n", ""))
    
for i in soup.find_all("td",class_='table-body__cell u-center-text'):# other teams ratings
    Matches.append(i.text)
    
Matches
```

```python
Points=[]

for i in soup.find_all("td",class_='rankings-block__banner--points'):# first place team ratings
    
    Points.append(i.text.replace("\n", ""))
    
for i in soup.find_all("td",class_='table-body__cell u-center-text'):# other teams ratings
    Points.append(i.text)
    
Points
```

```python
ratings=[]

for i in soup.find_all("td",class_='rankings-block__banner--rating u-text-right'):# first place team ratings
    
    ratings.append(i.text.replace("\n",""))
    
for i in soup.find_all("td",class_='table-body__cell u-text-right rating'):# other teams ratings
    ratings.append(i.text)
    
ratings
```

```python
print(len(Matches),len(Points),len(ratings))
```

```python
import pandas as pd
df = pd.DataFrame(Matches+Points+ratings)
df
```
