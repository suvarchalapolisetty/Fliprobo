# Model building


```python
df.info()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
import joblib
```

```python
X = df[['Director1', 'Director2', 'Duration', 'Year', 'Ratings', 'Metascore', 'Genre1', 'Certification']]
y_votes = df['Votes']
y_gross = df['Gross Collection']

# Split the data into training and testing sets
X_train, X_test, y_votes_train, y_votes_test, y_gross_train, y_gross_test = train_test_split(X, y_votes, y_gross, test_size=0.2, random_state=42)
```

```python
# Preprocess the data
# Perform one-hot encoding for categorical features
X_train_encoded = pd.get_dummies(X_train)
X_test_encoded = pd.get_dummies(X_test)

# Align the columns of the encoded test set with the training set
X_test_encoded = X_test_encoded.reindex(columns=X_train_encoded.columns, fill_value=0)

# Scale the numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_encoded)
X_test_scaled = scaler.transform(X_test_encoded)

```

```python
# Linear Regression
lr_model = LinearRegression()
lr_model.fit(X_train_scaled, y_votes_train)
lr_predictions = lr_model.predict(X_test_scaled)
lr_rmse = mean_squared_error(y_votes_test, lr_predictions, squared=False)
print("Linear Regression RMSE (Votes):", lr_rmse)

# Random Forest
rf_model = RandomForestRegressor(random_state=42)
param_grid = {'n_estimators': [100, 200, 300], 'max_depth': [None, 5, 10]}
grid_search = GridSearchCV(rf_model, param_grid, cv=5)
grid_search.fit(X_train_scaled, y_votes_train)
rf_model = grid_search.best_estimator_
rf_predictions = rf_model.predict(X_test_scaled)
rf_rmse = mean_squared_error(y_votes_test, rf_predictions, squared=False)
print("Random Forest RMSE (Votes):", rf_rmse)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR


# K-Nearest Neighbors
knn_model = KNeighborsRegressor()
knn_model.fit(X_train_scaled, y_votes_train)
knn_predictions = knn_model.predict(X_test_scaled)
knn_rmse = mean_squared_error(y_votes_test, knn_predictions, squared=False)
print("K-Nearest Neighbors RMSE (Votes):", knn_rmse)

# Support Vector Regression
svr_model = SVR()
svr_model.fit(X_train_scaled, y_votes_train)
svr_predictions = svr_model.predict(X_test_scaled)
svr_rmse = mean_squared_error(y_votes_test, svr_predictions, squared=False)
print("Support Vector Regression RMSE (Votes):", svr_rmse)


```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Gradient Boosting Regression
gb_model = GradientBoostingRegressor(random_state=42)
gb_model.fit(X_train_scaled, y_votes_train)
gb_predictions = gb_model.predict(X_test_scaled)
gb_rmse = mean_squared_error(y_votes_test, gb_predictions, squared=False)
print("Gradient Boosting Regression RMSE (Votes):", gb_rmse)
```

```python
import joblib

# Save the best model
best_model = gb_model  # Replace gb_model with the actual best model variable
joblib.dump(best_model, 'best_model_votes.pkl')
```
